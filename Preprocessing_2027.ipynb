* (10 points) จากข้อมูลที่ให้มา มีค่าว่างสูญหายมากน้อยแค่ไหน? และมีค่าผิดปกติ (outliers) และหรือข้อมูลรบกวน (noisy data) หรือไม่ในแต่ละตัวแปร?

```python
print(titanic.isnull().sum())
print(titanic.describe())

# อธิบายผลลัพธ์
# คอลัมน์ 'age' มีค่าว่างสูญหายจำนวนมาก
# คอลัมน์ 'cabin' มีค่าว่างสูญหายจำนวนมาก ซึ่งอาจหมายถึงข้อมูลไม่ครบถ้วน
# คอลัมน์ 'embarked' มีค่าว่างสูญหายเล็กน้อย
# จากการดูค่าสถิติเบื้องต้น พบว่าไม่มีค่าผิดปกติที่เห็นได้ชัดเจนในตัวแปรเชิงตัวเลข แต่ควรตรวจสอบข้อมูลเพิ่มเติมด้วยการสร้างกราฟเพื่อดูการกระจายของข้อมูล
```

* (10 points) จากข้อมูลที่ให้มา เตรียมข้อมูลให้พร้อมโดยการจัดการกับค่าว่างในข้อมูล ให้อธิบายวิธีที่ท่านเลือกใช้ พร้อมยกเหตุผลประกอบ

```python
# Age
# วิธี แทนที่ค่าว่างด้วยค่าเฉลี่ยของอายุ
# เหตุผล การแทนที่ด้วยค่าเฉลี่ยเป็นวิธีที่ง่ายและเหมาะสมสำหรับข้อมูลเชิงปริมาณ เนื่องจากไม่ทำให้การกระจายของข้อมูลเปลี่ยนแปลงมากนัก
titanic['age'].fillna(titanic['age'].mean(), inplace=True)

# Embarked
# วิธีมแทนที่ค่าว่างด้วยค่าที่พบมากที่สุด (mode)
# เหตุผล การแทนที่ด้วย mode เหมาะสมสำหรับข้อมูลเชิงคุณภาพ เนื่องจากไม่ทำให้การกระจายของข้อมูลเปลี่ยนแปลงมากนัก
titanic['embarked'].fillna(titanic['embarked'].mode()[0], inplace=True)

# Fare
# วิธี แทนที่ค่าว่างด้วยค่าเฉลี่ยของค่าโดยสาร
# เหตุผล การแทนที่ด้วยค่าเฉลี่ยเป็นวิธีที่ง่ายและเหมาะสมสำหรับข้อมูลเชิงปริมาณ เนื่องจากไม่ทำให้การกระจายของข้อมูลเปลี่ยนแปลงมากนัก
titanic['fare'].fillna(titanic['fare'].mean(), inplace=True)

# ตรวจสอบค่าว่างอีกครั้งหลังจากจัดการ
print(titanic.isnull().sum())
```

* แปลงข้อมูลเชิงคุณภาพให้เป็นตัวเลข \
    i. (10 points) ใช้ one-hot encoding ในการแปลงตัวแปร Embarked \
    ii. (10 points) แปลงตัวแปร Sex โดยให้เพศชายเป็น 0 และเพศหญิงเป็น 1 \
    iii. (10 points) สร้างตัวแปร FamilySize ซึ่งเป็นผลรวมของตัวแปร SibSp (number of siblings/spouses aboard) และ Parch (number of parents/children aboard) \
    iv. (10 points) สร้างตัวแปร IsAlone ซึ่งมีค่าเป็น 1 ถ้า FamilySize เป็น 0, และมีค่าเป็นศูนย์เมื่อเป็นอื่น

```python
# แปลงตัวแปร Embarked โดยใช้ one-hot encoding
embarked_encoded = pd.get_dummies(titanic['embarked'], prefix='embarked')
titanic = pd.concat([titanic, embarked_encoded], axis=1)
titanic.drop('embarked', axis=1, inplace=True)
```

```python
# แปลงตัวแปร Sex โดยให้เพศชายเป็น 0 และเพศหญิงเป็น 1
titanic['sex'] = titanic['sex'].map({'male': 0, 'female': 1})
```

```python
# สร้างตัวแปร FamilySize ซึ่งเป็นผลรวมของตัวแปร SibSp และ Parch
titanic['FamilySize'] = titanic['sibsp'] + titanic['parch']
```

```python
# สร้างตัวแปร IsAlone ซึ่งมีค่าเป็น 1 ถ้า FamilySize เป็น 0, และมีค่าเป็นศูนย์เมื่อเป็นอื่น
titanic['IsAlone'] = np.where(titanic['FamilySize'] == 0, 1, 0)
```

* ทำการ Normalize ข้อมูล \
    i. (10 points) ตัวแปร Fare และ Age ใช้ Min-Max Scaling \
    ii. (10 points) สร้างตัวแปรใหม่ Fare_Per_Person ด้วยการนำตัวแปร Fare มาหารด้วย FamilySize + 1 \
    iii. (10 points) สร้างตัวแปร AgeGroup ที่ระบุกลุ่มอายุของตัวอย่าง โดยแบ่งเป็นกลุ่มดังนี้ [0, 12), [12, 18), [18, 50) และ [50, 100)

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
titanic[['fare', 'age']] = scaler.fit_transform(titanic[['fare', 'age']])
```

```python
# สร้างตัวแปร Fare_Per_Person ด้วยการนำตัวแปร Fare มาหารด้วย FamilySize + 1
titanic['Fare_Per_Person'] = titanic['fare'] / (titanic['FamilySize'] + 1)
```

```python
# สร้างตัวแปร AgeGroup ที่ระบุกลุ่มอายุของตัวอย่าง โดยแบ่งเป็นกลุ่มดังนี้ [0, 12), [12, 18), [18, 50) และ [50, 100)
titanic['AgeGroup'] = pd.cut(titanic['age'], bins=[0, 12, 18, 50, 100], labels=['0-12', '12-18', '18-50', '50-100'])
```

* (10 points) นำเสนอการแจกแจงหรือทำความเข้าใจตัวแปรสำคัญในข้อมูล เช่น Age, Fare, and Pclass ด้วยการสร้างฮิสโตแกรมหรือแผนภาพกล่อง

```python
# Histograms for Age, Fare, and Pclass
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.hist(titanic['age'], bins=20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.subplot(1, 3, 2)
plt.hist(titanic['fare'], bins=20)
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Frequency')

plt.subplot(1, 3, 3)
plt.hist(titanic['pclass'], bins=3)
plt.title('Pclass Distribution')
plt.xlabel('Pclass')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# Age มีการกระจายตัวของอายุที่ค่อนข้างกว้าง มีกลุ่มอายุที่พบมากที่สุดอยู่ที่ประมาณ 20-30 ปี
# Fare มีการกระจายตัวของค่าโดยสารที่ค่อนข้างไม่สม่ำเสมอ มีกลุ่มค่าโดยสารที่พบมากที่สุดอยู่ที่ประมาณ 0-50
# Pclass แสดงให้เห็นถึงจำนวนผู้โดยสารในแต่ละชั้นที่นั่ง โดยส่วนใหญ่ผู้โดยสารอยู่ในชั้นที่ 3 มากที่สุด
```

* เมื่อเตรียมข้อมูลพร้อมแล้ว ให้หาความสัมพันธ์ของตัวแปรต่อไปนี้จากข้อมูล ซึ่งอาจนำเสนอด้วยสถิติและแผนภูมิหรือกราฟก็ได้: \
    i. (10 points) survived และ pclass \
    ii. (10 points) survived และ sex \
    iii. (10 points) survived และ age \
    iv. (10 points) จำนวนสมาชิกครอบครัวที่รอดชีวิต และ Fare

```python
# สร้างตาราง Pivot Table เพื่อดูความสัมพันธ์ระหว่าง survived และ pclass
pivot_table = titanic.pivot_table(index='survived', columns='pclass', values='name', aggfunc='count')
print(pivot_table)

# สร้างแผนภูมิแท่งเพื่อแสดงความสัมพันธ์ระหว่าง survived และ pclass
titanic.groupby(['pclass', 'survived'])['name'].count().unstack().plot(kind='bar', figsize=(10, 6))
plt.title('Relationship between Survived and Pclass')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.show()

# สรุปผล
# จากตาราง Pivot Table และแผนภูมิแท่ง พบว่าผู้โดยสารในชั้นที่ 1 มีอัตราการรอดชีวิตสูงกว่าผู้โดยสารในชั้นที่ 2 ซึ่งอาจเป็นเพราะผู้โดยสารในชั้นที่ 1 มีทรัพยากรและโอกาสในการรอดชีวิตสูงกว่า
```

```python
# สร้างตาราง Pivot Table เพื่อดูความสัมพันธ์ระหว่าง survived และ sex
pivot_table = titanic.pivot_table(index='survived', columns='sex', values='name', aggfunc='count')
print(pivot_table)

# สร้างแผนภูมิแท่งเพื่อแสดงความสัมพันธ์ระหว่าง survived และ sex
titanic.groupby(['sex', 'survived'])['name'].count().unstack().plot(kind='bar', figsize=(10, 6))
plt.title('Relationship between Survived and Sex')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()

# สรุปผล
# จากตาราง Pivot Table และแผนภูมิแท่ง พบว่าผู้หญิงมีอัตราการรอดชีวิตสูงกว่าผู้ชาย ซึ่งอาจเป็นเพราะผู้หญิงได้รับการช่วยเหลือก่อนในเหตุการณ์อับปาง
```

```python
# สร้าง scatter plot เพื่อดูความสัมพันธ์ระหว่าง survived และ age
plt.figure(figsize=(10, 6))
plt.scatter(titanic['age'], titanic['survived'])
plt.title('Relationship between Survived and Age')
plt.xlabel('Age')
plt.ylabel('Survived')
plt.show()

# สร้าง box plot เพื่อดูการกระจายตัวของอายุในแต่ละกลุ่ม survived
plt.figure(figsize=(10, 6))
titanic.boxplot(column='age', by='survived')
plt.title('Age Distribution by Survived')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.show()

# คำนวณค่าสถิติพื้นฐานของอายุในแต่ละกลุ่ม survived
print(titanic.groupby('survived')['age'].describe())

# สรุปผล
# จาก scatter plot และ box plot พบว่าอายุมีความสัมพันธ์กับการรอดชีวิตอยู่บ้าง
# โดยผู้โดยสารที่มีอายุมากกว่ามีแนวโน้มที่จะรอดชีวิตน้อยกว่าผู้โดยสารที่มีอายุต่ำกว่า นอกจากนี้ยังพบว่าผู้โดยสารที่รอดชีวิตมีอายุเฉลี่ยสูงกว่าผู้โดยสารที่เสียชีวิต
```

```python
# สร้าง scatter plot เพื่อดูความสัมพันธ์ระหว่าง FamilySize และ Fare
plt.figure(figsize=(10, 6))
plt.scatter(titanic['FamilySize'], titanic['fare'])
plt.title('Relationship between FamilySize and Fare')
plt.xlabel('FamilySize')
plt.ylabel('Fare')
plt.show()

# คำนวณค่าสถิติพื้นฐานของ Fare ในแต่ละกลุ่ม FamilySize
print(titanic.groupby('FamilySize')['fare'].describe())

# สร้าง box plot เพื่อดูการกระจายตัวของ Fare ในแต่ละกลุ่ม FamilySize
plt.figure(figsize=(10, 6))
titanic.boxplot(column='fare', by='FamilySize')
plt.title('Fare Distribution by FamilySize')
plt.xlabel('FamilySize')
plt.ylabel('Fare')
plt.show()

# สรุปผล
# จาก scatter plot และ box plot พบว่าขนาดของครอบครัวมีความสัมพันธ์กับค่าโดยสารอยู่บ้าง
# โดยทั่วไปแล้ว ผู้โดยสารที่มีครอบครัวใหญ่จะมีค่าโดยสารสูงกว่าผู้โดยสารที่เดินทางคนเดียว แต่ก็มีบางกรณีที่ผู้โดยสารที่เดินทางคนเดียวมีค่าโดยสารสูงกว่าผู้โดยสารที่มีครอบครัวใหญ่ด้วย
```

```python
# สร้างตารางสรุปข้อมูลเพื่อหาคุณลักษณะของผู้โดยสารเรือไททานิค และศึกษาเกี่ยวกับการเสียชีวิตของผู้โดยสาร
summary_table = titanic.groupby(['sex', 'pclass', 'AgeGroup'])['survived'].agg(['sum', 'count', 'mean'])
print(summary_table)

# สรุปผลที่ได้จากตาราง
# ผู้โดยสารหญิงมีอัตราการรอดชีวิตสูงกว่าผู้โดยสารชาย
# ผู้โดยสารชั้น 1 มีอัตราการรอดชีวิตสูงกว่าผู้โดยสารชั้น 2 และ 3
# ผู้โดยสารในกลุ่มอายุ 0-12 มีอัตราการรอดชีวิตสูงกว่ากลุ่มอายุอื่นๆ
```
